```python
import numpy as np
import pandas as pd
import time
groundtruth = '../dataset/open_peer_review_v2/peer_review/translated_groundtruth.csv'
peer_reciew = '../dataset/open_peer_review_v2/peer_review/peer_review_forPG3.csv'
gDF = pd.read_csv(groundtruth)
rDF = pd.read_csv(peer_reciew)
```

### PG1+PC1

```python
%run ../tools/PG1.py
%run ../tools/PC1.py
modelPG1 = PG1(gDF,rDF)
modelPC1 = PC1(gDF,rDF)
```

```python
start = time.time()
pg1_fit = modelPG1.fit([1.54705059, 2.64350654, 62.67778188, 78.70005719, 1.33529373])
elapsed_time = time.time() - start
pg1_eap = pg1_fit.summary()['summary'][:,0]
pg1_estimated = pg1_eap[0:modelPG1.userNum]
print(elapsed_time, np.corrcoef(modelPG1.grade, pg1_estimated)[0,1])

start = time.time()
pc1_fit = modelPC1.fit([-1.83033443, 2.74556604, 65.2612221, 74.74509713])
elapsed_time = time.time() - start
pc1_eap = pc1_fit.summary()['summary'][:,0]
pc1_estimated = pc1_eap[0:modelPC1.userNum]
print(elapsed_time, np.corrcoef(modelPG1.grade, pc1_estimated)[0,1])
```

```python
for i in xrange(100):
    rate = i*0.01
    ensemble = pg1_estimated*(1-rate) + pc1_estimated*rate
    cor = np.corrcoef(modelPG1.grade, ensemble)[0,1]
    print(rate, cor)
```

```python
mrate = 0.32
ensemble = pg1_estimated*(1-mrate) + pc1_estimated*mrate
np.corrcoef(modelPG1.grade, ensemble)[0,1]
```
